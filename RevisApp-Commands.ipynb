```python
import os
import subprocess
import sys
from IPython.display import display, HTML, Markdown

# Verificar Node.js e npm
print("üîç Verificando ambiente...\n")

try:
    node_version = subprocess.run(["node", "--version"], capture_output=True, text=True, shell=True)
    print(f"‚úÖ Node.js: {node_version.stdout.strip()}")
except:
    print("‚ùå Node.js n√£o encontrado")

try:
    npm_version = subprocess.run(["npm", "--version"], capture_output=True, text=True, shell=True)
    print(f"‚úÖ npm: {npm_version.stdout.strip()}")
except:
    print("‚ùå npm n√£o encontrado")

# Verificar diret√≥rio do projeto
print(f"\nüìÅ Diret√≥rio atual: {os.getcwd()}")
print(f"üìÅ Arquivos principais:")
for file in ["package.json", "vite.config.ts", "capacitor.config.ts"]:
    exists = "‚úÖ" if os.path.exists(file) else "‚ùå"
    print(f"   {exists} {file}")
```
